# Condicionales

```python
edad = 37

if edad == 25:
    print("Tiene 25 años.")
else:
    print(f"No tiene 25 años. Tiene {edad} años.")
```

```python
edad = 40
altura = 150

if edad > 18 and altura > 160:
    print("Puede entrar al parque")
else:
    print("No puede entrar al parque")
```

```python
edad = 15
altura = 170

if edad >= 18 and altura > 160:
    print("Puede entrar al parque")
elif edad > 12 and edad < 18 and altura > 160:
    print("Puede entrar acompañado/a al parque")
else:
    print("No puede entrar al parque")
```

# Bucles

## FOR

```python
lista = [0,1,2,3,4]
```

```python
for i in lista:
    print(i)
```

```python
for i in range(0,5):
    print(i)
```

```python
for i in range(1,4):
    print(f"Tabla del {i}")
    for j in range(1,11):
        print(f"{i} * {j} = {i*j}")
```

### Bucles con strings

```python
palabra = "perro"
```

```python
for i in palabra:
    print(i)
```

```python
acum = ""
for i in palabra:
    acum += i
    print(acum)    
```

## While

```python
i = 0
while i < 5:
    print(i)
    i += 1
```
